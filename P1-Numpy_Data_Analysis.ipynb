# Data Analysis with Numpy

## Introduction

In this project, we will perform statistical analysis on the Seattle weather dataset using NumPy. This dataset includes daily weather observations from Seattle, such as temperature, precipitation, and wind. The objective is to calculate basic statistical measures and visualize the results.

## Steps
1. Load the Dataset
2. Basic Statistics
3. Visualizations

### Environment Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Load Data Set
We begin by loading the Seattle weather dataset using Pandas. This allows us to read the CSV file and display the first few rows to understand its structure.

```python
data = pd.read_csv(r'seattle-weather.csv')

data.head()
```

### Exclude String Columns

The dataset contains both numerical and string columns. For our numerical analysis, we need to exclude the string columns. This step ensures that only numerical data is used for calculating statistical measures.

```python
num_data = data[['precipitation', 'temp_max', 'temp_min', 'wind']]
num_data
```

### Convert Numerical Data to Numpy Arrays

After excluding the string columns, we convert the remaining numerical data into NumPy arrays. This allows us to leverage NumPy's efficient array operations for statistical analysis.

```python
data_array = num_data.to_numpy()

data_array
```

```python
data_array.shape
```

## 2. Basic Statistics

### Calculate the Mean of the Numerical Data 

The mean provides the average value of each column.

```python
mean_value = np.mean(data_array, axis = 0)
print(f"Mean Values:\t{mean_value}")
```

### Calculate the Median of the Numerical Data

The median is the middle value that separates the higher half from the lower half of the data.

```python
median_value = np.median(data_array, axis = 0)
print(f"Median Values:\t{median_value}")
```

### Calculate the Standard Deviation

The standard deviation measures the amount of variation or dispersion in the dataset.

```python
std_deviation = np.std(data_array, axis = 0)
print(f"Standard Deviation Values:\t{std_deviation}")
```

### Calculate the Correlation Matrix

Correlation measures the linear relationship between two variables.

```python
corr_matrix = np.corrcoef(data_array, rowvar = False) 
print(f"Correlation Matrix Values:\n{corr_matrix}")
```

## 3. Visualizations

### Plotting of Mean Value Data

Plotting the mean values for each column.

```python
plt.figure(figsize = (10, 5))
plt.plot(mean_value, marker = 'o', linestyle = '-', color = 'r')
plt.title('Mean Values of Weather Data Columns')
plt.xlabel('Column Index')
plt.ylabel('Mean Value')
plt.show()
```

### Plotting of Median Value Data

Plotting the median values for each column.

```python
plt.figure(figsize = (10, 5))
plt.plot(median_value, marker = 'o', linestyle = '-', color = 'g')
plt.title('Median Values of Weather Data Columns')
plt.xlabel('Column Index')
plt.ylabel('Median Value')
plt.show()
```

### Plotting of Standard Deviation Value Data

Plotting the standard deviation for each column.

```python
plt.figure(figsize = (10, 5))
plt.plot(std_deviation, marker = 'o', linestyle = '-', color = 'b')
plt.title('Standard Deviation of Weather Data Columns')
plt.xlabel('Column Index')
plt.ylabel('Standard Deviation')
plt.xticks(range(len(num_data.columns)), num_data.columns, rotation = 45)
plt.show()
```

### Plotting of Correlation Matrix Data as Heatmap

Visualizing the correlation matrix using a heatmap.

```python
plt.figure(figsize=(8, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar(label='Correlation Coefficient')
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(num_data.columns)), num_data.columns, rotation=45)
plt.yticks(range(len(num_data.columns)), num_data.columns)
plt.show()
```
