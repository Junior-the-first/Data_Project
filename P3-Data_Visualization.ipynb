# Data Visualization
## Introduction

In this project, we will showcase data visualization techniques using Matplotlib and Seaborn. We will visualize the Seattle weather data to understand the distribution and relationships between different weather parameters.

## Steps

1. Load the Cleaned Dataset
2. Histogram
3. Scatter Plot

### Environment Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Load Data Set

We start by loading the cleaned Seattle weather dataset.

```python
data = pd.read_csv(r'cleaned_seattle-weather.csv')

data.head()
```

# Set the style for the plots

We set the style for the plots using 'sns.set(style="whitegrid")'. 
This style adds a grid to the background of the plots, making them easier to read and interpret.

```python
sns.set(style="whitegrid")
```

## 2. Histogram

We create a histogram to show the distribution of maximum temperatures. This helps in understanding the spread and frequency of temperature values.

### Histogram of Maximum Temperature

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['temp_max'], bins=20, kde=True)
plt.title('Distribution of Maximum Temperature')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.show()
```

## 3. Scatter Plot

We visualize the relationship between maximum temperature and precipitation using a scatter plot. This helps in understanding how temperature and precipitation are related.

### Scatter Plot of Max Temperature vs Precipitation

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp_max', y='precipitation', data=data)
plt.title('Max Temperature vs Precipitation')
plt.xlabel('Max Temperature (C)')
plt.ylabel('Precipitation (mm)')
plt.show()
```
